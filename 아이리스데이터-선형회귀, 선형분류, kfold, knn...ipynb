```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import *

from sklearn import datasets 
from sklearn.model_selection import *
from sklearn.linear_model import *

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
iris = datasets.load_iris() 
iris## 딕셔너리
```

```python
iris.feature_names  #열이름
```

```python
X_all = iris.data ## numpy
X_all[:3]  ##0행, 1행, 2행을 추출
```

```python
X = X_all[:, 0]  ##모든행의 0열 - sepal length
y = X_all[:, 2]  ##모든행의 2열 - petal length
print(X[0:5])    
print(y[0:5])
```

```python
plt.scatter(X, y, marker='o') 
plt.xlabel("Sepal length") ; plt.ylabel("Petal length")
```

```python
X_train, X_test, y_train, y_test \
      = train_test_split(X, y, test_size=0.2, random_state=7)
plt.scatter(X_train, y_train, marker='o') 
plt.xlabel("Sepal length") ; plt.ylabel("Petal length")
```

```python
plt.scatter(X_test, y_test, marker='o') 
plt.xlabel("Sepal length") ; plt.ylabel("Petal length")
```

```python
linr = LinearRegression()
linr.fit(X_train.reshape(-1,1), y_train)  ## 선형회귀 모델을 적용

# LinearRegression().fit(X_train.reshape(-1,1), y_train)

print("Score : {:.3f}".format(linr.score(X_test.reshape(-1,1), y_test)))
print(linr.score(X_test.reshape(-1,1), y_test))  ## 검증 신뢰도
print(linr.coef_)  
print(linr.intercept_)
```

```python
plt.scatter(X, y, marker='o') 
plt.xlabel("Sepal length")
plt.ylabel("Petal length")

XX = np.linspace(4, 8, 200) ; 
plt.plot(XX, linr.coef_ * XX + linr.intercept_, "k-")
```

## 선형분류

```python
iris.feature_names  ##열이름, 피처, 독립변수
```

```python
iris.target_names  ## 꽃의 종류 - 종속변수
# 'setosa' : 0 , 'versicolor' : 1, 'virginica' :2
```

```python
X, y = iris.data, iris.target
X.shape, y.shape
```

```python
X[0:4], y[0:4],  X[145:], y[145:]  ## numpy
```

```python
# 'setosa' : 0   'Nonsetosa' : 1
X2 = X[:, :2]    ##0열, 1열만 - 꽃받침의 길이와 넓이를 이용해서 종류를 예측
y2 = y.copy()              
y2[(y2==2)] = 1
y2
```

```python
markers = ['o', '+', '^']
for i in range(3):
    xs = X2[:, 0][y == i]
    ys = X2[:, 1][y == i]
    plt.scatter(xs, ys, marker=markers[i])
plt.legend(iris.target_names)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X2, y2, test_size=0.2)
```

```python
markers = ['o', '+', '^']
for i in range(3):
    xs = X_train[:, 0][y_train == i]
    ys = X_train[:, 1][y_train == i]
    plt.scatter(xs, ys, marker=markers[i])
binary_names = ['setosa', 'non-setosa']
plt.legend(binary_names)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
```

```python
from sklearn.linear_model import SGDClassifier
clf = SGDClassifier(max_iter=1000)
clf.fit(X_train, y_train)
a = clf.coef_[0,0]  
b = clf.coef_[0,1]  
c = clf.intercept_  ##y절편
a, b, c, clf.coef_, clf.intercept_
```

```python
markers = ['o', '+', '^']
for i in range(3):
    xs = X_train[:, 0][y_train == i]
    ys = X_train[:, 1][y_train == i]
    plt.scatter(xs, ys, marker=markers[i])
binary_names = ['setosa', 'non-setosa']
plt.legend(binary_names)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

XX = np.linspace(4, 8, 40)
plt.plot(XX, (-a/b)*XX - c/b, "k-", linewidth=2)
```

```python
clf = SGDClassifier(loss='perceptron', max_iter=1000)
clf.fit(X_train, y_train)
a = clf.coef_[0,0]  
b = clf.coef_[0,1]  
c = clf.intercept_  ##y절편
a, b, c, clf.coef_, clf.intercept_
```

```python
markers = ['o', '+', '^']
for i in range(3):
    xs = X_train[:, 0][y_train == i]
    ys = X_train[:, 1][y_train == i]
    plt.scatter(xs, ys, marker=markers[i])
binary_names = ['setosa', 'non-setosa']
plt.legend(binary_names)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

XX = np.linspace(4, 8, 40)
plt.plot(XX, (-a/b)*XX - c/b, "k-", linewidth=2)
```

```python
clf.predict([[6.5, 2.5]]) ##nonsetosa ## 모델 확정 후 실제 활용한 테스트
```

## knn

```python
from sklearn.neighbors import KNeighborsClassifier
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

knn = KNeighborsClassifier(n_neighbors=3) ## KNN - 분류
knn.fit(X_train, y_train)
print("kNN score: {:.2f}".format(knn.score(X_test, y_test)))

clf = SGDClassifier()   ## 확률적 경사하강법 - 선형 분류
clf.fit(X_train, y_train)
print("Linear Reg score: {:.2f}".format(clf.score(X_test, y_test)))
```

```python
X.shape
```

```python
for i in range(1,100,10):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    print("K가", i, "일때 정확도: {:.2f}".format(knn.score(X_test, y_test)))
```

```python
cross_val_score(knn, X, y, cv=5).round(4)
```

```python
cross_val_score(knn, X, y, cv=5).mean().round(4)
```

```python
cross_val_score(knn, X, y, scoring = "r2", cv=5).mean().round(4)
```

```python
cross_val_score(clf, X, y, cv=5).mean().round(4)
```
