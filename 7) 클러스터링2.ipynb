```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
#주피터노트북의 명령어 - 차트가 바로 보이게 하는 것
%config InlineBackend.figure_format = 'retina'
## 맥에서는 꼭 적어줘야하는 것
from sklearn.datasets import make_moons
from sklearn.cluster import *
```

```python
X, y = make_moons(n_samples=3000, noise=0.05, random_state=11)
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
X, y = make_moons(n_samples=300, noise=0.05, random_state=11)
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
predict = kmeans.fit_predict(X) 
##X데이터를 kmean를 기준으로 그룹번호를 부여하라
```

```python
#kmeans = KMeans(n_clusters= 2).fit(X)
kmeans2 = kmeans.fit(X)
centers = kmeans.cluster_centers_
centers
```

```python
plt.scatter(X[:,0], X[:,1],c=predict)
```

```python
dbscan = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
##eps : 묶는 반경 , min_samples : 1개의 반경에 들어가는 샘플의 최소 갯수
predict = dbscan.fit_predict(X)
plt.scatter(X[:,0], X[:,1],c=predict)
```

```python
dbscan = DBSCAN(eps=0.2, min_samples=10, metric='euclidean')
##eps : 묶는 반경 , min_samples : 1개의 반경에 들어가는 샘플의 최소 갯수
predict = dbscan.fit_predict(X)
plt.scatter(X[:,0], X[:,1],c=predict)
```

```python
dbscan = DBSCAN(eps=0.1, min_samples=5, metric='euclidean')
predict = dbscan.fit_predict(X)
plt.scatter(X[:,0], X[:,1],c=predict)
```

```python
dbscan = DBSCAN(eps=0.1, min_samples=10, metric='euclidean')
predict = dbscan.fit_predict(X)
plt.scatter(X[:,0], X[:,1],c=predict)
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=10, metric='euclidean')
predict = dbscan.fit_predict(X)
plt.scatter(X[:,0], X[:,1],c=predict)
```

n_samples : 표본 데이터의 수, 디폴트 100

n_features : 독립 변수의 수, 디폴트 20 (속성,열의 갯수)

centers : 생성할 클러스터의 수 혹은 중심점,
          [n_centers, n_features] 크기의 배열. 디폴트 3

cluster_std: 클러스터의 표준 편차, 디폴트 1.0

center_box:생성할 클러스터의 바운딩 박스(bounding box), 디폴트 (-10.0, 10.0))

```python
X,y = make_blobs(n_features=2, n_samples=1000, cluster_std=.5)
plt.scatter(X[:,0], X[:,1],s=10)
```
